Este notebook contiene el an谩lisis exploratorio del dataset de predicci贸n de demanda de Mango.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

# Configuraci贸n
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

pd.set_option('display.max_columns', 100)
```

## 1. Carga de Datos


```python
# Cargar datos
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')
sample_submission = pd.read_csv('../data/sample_submission.csv')

print(" Dimensiones:")
print(f"Train: {train_df.shape}")
print(f"Test: {test_df.shape}")
print(f"\n Primeras filas:")
train_df.head()
```

## 2. An谩lisis de la Variable Target (Production)


```python
print(" Estad铆sticas de Production:")
print(train_df['Production'].describe())

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histograma
axes[0].hist(train_df['Production'], bins=50, edgecolor='black', alpha=0.7)
axes[0].set_title('Distribuci贸n de Production')
axes[0].set_xlabel('Production')
axes[0].set_ylabel('Frecuencia')

# Boxplot
axes[1].boxplot(train_df['Production'])
axes[1].set_title('Boxplot de Production')

# Log scale
axes[2].hist(np.log1p(train_df['Production']), bins=50, edgecolor='black', alpha=0.7, color='coral')
axes[2].set_title('Distribuci贸n log(Production)')

plt.tight_layout()
plt.show()
```
