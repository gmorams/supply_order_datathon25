```python
import numpy as np
def asymmetric_score(real, pred):
    diff = np.abs(pred - real)
    penalty = np.where(pred < real, 2, 1)
    loss = diff * penalty

    # Normalitzar per fer un score entre 0 i 100
    norm_loss = loss / (real + 1e-6)
    score = 100 / (1 + norm_loss.mean())
    return score
```

```python
import pandas as pd
data = pd.read_csv("train.csv", sep=';', encoding='utf-8')
```

```python
def prep_test(val_df):
  val_demand = (
      val_df.groupby("ID")["weekly_demand"]
      .sum()
      .reset_index()
      .rename(columns={"weekly_demand": "demand"})
  )
  val_static = val_df.drop_duplicates("ID").drop(columns=["weekly_demand"])
  val_final = val_static.merge(val_demand, on="ID", how="inner")
  validation = val_final.drop(columns=["num_week_iso","weekly_sales","Production", "year"])
  return validation
```

```python
def split(data):# Dividim en train i test
  ids = data["ID"].unique()
  print(len(ids), "IDs Ãºnics")
  from sklearn.model_selection import train_test_split

  train_ids, val_ids = train_test_split(
      ids,
      test_size=0.20,
      random_state=42
  )

  train_df = data[data["ID"].isin(train_ids)].reset_index(drop=True)
  val_df   = data[data["ID"].isin(val_ids)].reset_index(drop=True)

  train_df.shape, val_df.shape
  test_df = prep_test(val_df)
  return train_df, test_df
```

```python
train, test = split(data)
```

```python
def preprocess(df):
    df = df.copy()

    # --- RGB ---
    rgb = df["color_rgb"].str.extract(r"(\d+),\s*(\d+),\s*(\d+)").astype(float)
    df[["R","G","B"]] = rgb
    df = df.drop(columns=["color_rgb"])

    # --- One-hot encoding ---
    categorical_cols = [
        "aggregated_family","family","category","fabric","color_name",
        "length_type","silhouette_type","waist_type","neck_lapel_type",
        "sleeve_length_type","heel_shape_type","toecap_type",
        "woven_structure","knit_structure","print_type",
        "archetype","moment"
    ]
    df = pd.get_dummies(df, columns=categorical_cols)

    return df
```

```python
from numpy.linalg import norm

def cosine_sim(a, b):
    return np.dot(a, b) / (norm(a) * norm(b) + 1e-9)
```

```python
def predict_one(test_emb, train_embs, train_demands, k=10):
    sims = np.array([cosine_sim(test_emb, e) for e in train_embs])
    topk = sims.argsort()[-k:]  # agafem els mÃ©s semblants
    return train_demands[topk].mean()  # predicciÃ³
```

```python
import numpy as np
import pandas as pd

def evaluate_predictions(df, demand_col="demand", pred_col="prediction", under_penalty=3.0):
    """
    Calcula un score entre 0 i 100 penalitzant mÃ©s quan la predicciÃ³ Ã©s inferior a la demanda real.
    """
    real = df[demand_col].values
    pred = df[pred_col].values

    errors = np.abs(pred - real)

    # Penalitzar mÃ©s els casos on quedem curts
    penalties = np.where(pred < real, under_penalty, 1)

    weighted_errors = errors * penalties

    mean_error = weighted_errors.mean()

    # Convertir error en score (com menys error, mÃ©s score)
    # Ajustar divisor 50 segons magnitud de les teves demandes
    score = max(0, 100 * (1 - mean_error / 50))

    return {
        "mean_error": mean_error,
        "score": round(score, 2),
        "pct_under": float((pred < real).mean() * 100),
        "pct_over": float((pred >= real).mean() * 100)
    }
```

```python
# totes les columnes menys la target
feature_cols = [c for c in train_cat.columns if c != "demand"]

# categoricals = columnes objecte (strings)
cat_cols = [c for c in feature_cols if train_cat[c].dtype == "object"]

# numeric = la resta (inclou emb_0...emb_511)
num_cols = [c for c in feature_cols if c not in cat_cols]
```

```python
X_train = train_cat[feature_cols]
y_train = train_cat["demand"]

X_test = test_cat[feature_cols]
y_test = test_cat["demand"]
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(
    iterations=700,
    learning_rate=0.06,
    depth=10,
    loss_function="MAE",
    verbose=200
)

model.fit(
    X_train, y_train,
    cat_features=cat_cols
)
```

```python
y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

print("MAE:", mae)
print("RMSE:", rmse)
```
